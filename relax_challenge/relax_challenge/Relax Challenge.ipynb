```python
user2 = engagement[engagement['user_id'] == 2]


user2_times = [datetime.strptime(string, '%Y-%m-%d %H:%M:%S') for string in list(user2['time_stamp'])]

def count_distance(target_time, time_list, time_difference):
    count = 0 
    for time in time_list:
        if abs(target_time - time) < time_difference:
            count += 1
    return count
```

```python
engagement['time_stamp'] = engagement['time_stamp'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
```
