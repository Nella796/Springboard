/* Q7: Produce a list of all members who have used a tennis court.
Include in your output the name of the court, and the name of the member
formatted as a single column. Ensure no duplicate data, and order by
the member name. */

SELECT DISTINCT B.memid, CONCAT(M.firstname, ' ' ,M.surname) AS member, F.name AS facility
FROM Bookings B
LEFT JOIN Facilities F ON F.facid = B.facid 
LEFT JOIN Members M ON B.memid = M.memid
WHERE F.name LIKE '%Tennis Court%'
ORDER BY member


/* Q8: Produce a list of bookings on the day of 2012-09-14 which
will cost the member (or guest) more than $30. Remember that guests have
different costs to members (the listed costs are per half-hour 'slot'), and
the guest user's ID is always 0. Include in your output the name of the
facility, the name of the member formatted as a single column, and the cost.
Order by descending cost, and do not use any subqueries. */

SELECT B.memid, CONCAT(M.firstname, ' ' ,M.surname) AS member, F.name AS facility, F.membercost, F.guestcost, 
CASE WHEN B.memid = 0 THEN F.guestcost
ELSE F.membercost END AS Cost
FROM Bookings B
LEFT JOIN Facilities F ON F.facid = B.facid 
LEFT JOIN Members M ON B.memid = M.memid
WHERE B.starttime LIKE '%2012-09-14%'
AND CASE WHEN B.memid = 0 THEN F.guestcost
ELSE F.membercost END > 30
ORDER BY CASE WHEN B.memid = 0 THEN F.guestcost
ELSE F.membercost END DESC

```python
''' Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! */'''


def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT name, revenue FROM

(
SELECT name, SUM(Cost) AS revenue FROM

(SELECT B.bookid, F.Facid, F.name,
CASE WHEN B.memid = 0 THEN F.guestcost
ELSE F.membercost END AS Cost FROM Bookings B
INNER JOIN Facilities F ON B.facid = F.facid)

AS query1

GROUP BY name
)

AS query2

WHERE revenue < 1000
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
'''Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order '''

import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT m1.surname, m1.firstname, m2.firstname, m2.surname
FROM Members AS m1
LEFT JOIN Members AS m2 ON m1.memid = m2.recommendedby
WHERE m1.surname != 'GUEST'
ORDER BY m1.surname , m1.firstname
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
''' Q12: Find the facilities with their usage by member, but not guests '''
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT M.firstname, M.surname, B.memid, COUNT(B.memid), B.facid, F.name FROM Bookings B
INNER JOIN Facilities F ON F.facid = B.facid
INNER JOIN Members M ON M.memid = B.memid
WHERE B.memid != 0
GROUP BY F.name, B.memid
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
''' Q13: Find the facilities usage by month, but not guests '''

def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT COUNT(B.memid), B.facid, F.name, strftime('%m', starttime) FROM Bookings B
INNER JOIN Facilities F ON F.facid = B.facid
INNER JOIN Members M ON M.memid = B.memid
WHERE B.memid != 0
GROUP BY F.name, strftime('%m', starttime)
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
